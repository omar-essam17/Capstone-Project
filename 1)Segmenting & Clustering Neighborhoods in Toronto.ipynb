```python
# importing Libraries
import pandas as pd
import requests
```

```python
# save the wikipedia page url in url variable
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

```python
# get the content of the page using requests
page = requests.get(url).content
```

```python
# read the html file of the page 
df = pd.read_html(page, header=0)[0]
df.head(10)
```

```python
# Only process the cells that have an assigned borough. Ignore cells with a borough that is Not assigned.
df_tor = df[df.Borough != 'Not assigned']
```

```python
df_tor.head(10)
```

```python
# drop duplicates values from the data frame
df_tor = df_tor.drop_duplicates()
df_tor.head(10)
```

```python
# rename the columns into PostalCode & Neighborhood
df_tor.rename(columns = {"Postal Code":"PostalCode", "Neighbourhood":"Neighborhood"} , inplace = True)
df_tor.reset_index(inplace=True, drop=True) # rest the index
df_tor.head()
```
